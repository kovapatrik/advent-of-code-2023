```python
import numpy as np
from collections import defaultdict
```

```python
content = open('input.txt').read().split('\n')
```

```python
content
```

```python
rows = len(content)
cols = len(content[0])
```

```python
# doublerows 
drows =[r for r in np.arange(rows) if content[r].find('#')==-1]
print(drows)
```

```python
dcols = list()
for c in np.arange(cols):
    found = 0
    for r in np.arange(rows):
        if '#' in content[r][c]:
            found = 1
            break
    if found == 0:
        dcols.append(c)
```

```python
dcols
```

```python
counter = 0
galaxis = defaultdict()
for i,line in enumerate(content):
    for j in np.arange(line.count('#')):
        if j==0:
            galaxis[counter] = [i,line.find('#')]
            counter+=1
            prev_val = line.find('#')
        else:
            galaxis[counter] = [i,line[prev_val+1:].find('#')+prev_val+1]
            counter+=1
            prev_val = line[prev_val+1:].find('#') + prev_val + 1
        
```

```python
galaxis
```

```python
from itertools import combinations
```

```python
distance = 0
for k in combinations(galaxis,2):
    rx = range(min(galaxis[k[0]][0],galaxis[k[1]][0]),max(galaxis[k[0]][0],galaxis[k[1]][0]))
    ry = range(min(galaxis[k[0]][1],galaxis[k[1]][1]),max(galaxis[k[0]][1],galaxis[k[1]][1]))
    rmodifier = sum([rx.count(dr)>0 for dr in drows ])
    cmodifier = sum([ry.count(dc)>0 for dc in dcols ])
#     print(k,rx,ry,abs(rx.stop-rx.start)+abs(ry.stop-ry.start)+rmodifier+cmodifier)
    distance+=(abs(rx.stop-rx.start)+abs(ry.stop-ry.start)+rmodifier*999999+cmodifier*999999)
```
