```python
from collections import defaultdict
import numpy as np
import itertools
```

```python
content = open('input.txt').readlines()
```

```python
[print(l) for l in content]
```

```python
movein= {'F':'NW', 'J':'SE', '-':'EW', '|':'NS', '7':'EN', 'L':'SW','S':'NSEW'} 
moveout= {'F':'SE', 'J':'NW', '-':'EW', '|':'NS', '7':'WS', 'L':'NE','S':'NSEW'}   
direction = {
    'N':[-1,0],
    'S':[1,0],
    'E':[0,1],
    'W':[0,-1]
}
```

```python
coordinatesofS=[0,0]
for i,line in enumerate(content):
    if 'S' in line:
        coordinatesofS = [i, line.find('S')]
        break
```

```python
coordinatesofS
```

```python
actual_coordinates = coordinatesofS
prev_coord = [-1, -1]
actchar = content[actual_coordinates[0]][actual_coordinates[1]] 
nextchar = ''
steps = 0
minrow = 0
maxrow = len(content)
mincol = 0
maxcol = len(content[0])
firs_step = 0
cycle = 0
coords = defaultdict()
coords[tuple(coordinatesofS)] = steps
while ('S' not in nextchar):
    cycle +=1
    found_nextstep = 0
    for d in direction.keys():
#         print(d)
        if found_nextstep==1:
#             print('breakactive')
            break
#             print('breakactive2')
        nextcoord = [direction[d][0] + actual_coordinates[0],direction[d][1] + actual_coordinates[1]] # next step
#         print(nextcoord)
        if (minrow <= nextcoord[0] < maxrow) & (mincol <= nextcoord[1] <maxcol): # next step inside map
            if  not ((nextcoord[0]-prev_coord[0]) ==0 ) & ((nextcoord[1]-prev_coord[1]) == 0): #next step not previous
                nextchar = content[nextcoord[0]][nextcoord[1]] 
                if nextchar in movein.keys(): # if part of pipeline
#                     print(actchar, nextchar, actual_coordinates,nextcoord,steps)
                    if len(set(movein[nextchar]).intersection(set(moveout[actchar])).intersection(set(d)))>0:
                        steps+=1
#                         print(actchar, nextchar, actual_coordinates,nextcoord,steps)
                        prev_coord = actual_coordinates.copy()
                        actchar = nextchar
                        coords[tuple(actual_coordinates)] = steps
                        actual_coordinates = nextcoord.copy()   
                        found_nextstep = 1
                        
#         else:
#             print('outside map',d)
        
#     if cycle>25:
#         print('debugbreak')
#         break
    if steps%1000==0:
        print(steps)
```

```python
counter = 0
insider = defaultdict()
prev = list(coords.keys())[0]
for k in list(coords.keys())[1:]:
    direction = [k[0]-prev[0],k[1]-prev[1]]
    lefty =[-direction[1],direction[0]]
    multiplier = 1
    innertile = [k[0]+lefty[0]*multiplier, k[1]+lefty[1]*multiplier]
#     print(innertile, (coords.get(tuple(innertile)) is None), (insider.get(tuple(innertile)) is None) )
    while(coords.get(tuple(innertile)) is None):
        if insider.get(tuple(innertile)) is None:
            counter +=1
            print(k,innertile)
            content[innertile[0]] = content[innertile[0]][:innertile[1]] + 'X' + content[innertile[0]][innertile[1]+1:]
            insider[tuple(innertile)] = counter
        
        multiplier +=1
        innertile = [k[0]+lefty[0]*multiplier, k[1]+lefty[1]*multiplier]
    prev = k
prev = list(coords.keys())[0]
for k in list(coords.keys())[::-1][:-1]:
    direction = [k[0]-prev[0],k[1]-prev[1]]
    lefty =[direction[1],-direction[0]]
    multiplier = 1
    innertile = [k[0]+lefty[0]*multiplier, k[1]+lefty[1]*multiplier]
#     print(innertile, (coords.get(tuple(innertile)) is None), (insider.get(tuple(innertile)) is None) )
    while(coords.get(tuple(innertile)) is None):
        if insider.get(tuple(innertile)) is None:
            counter +=1
            print(k,innertile)
            content[innertile[0]] = content[innertile[0]][:innertile[1]] + 'X' + content[innertile[0]][innertile[1]+1:]
            insider[tuple(innertile)] = counter
        
        multiplier +=1
        innertile = [k[0]+lefty[0]*multiplier, k[1]+lefty[1]*multiplier]
    prev = k
print(counter)    
```
